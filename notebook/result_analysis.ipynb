```python
import numpy as np
from math import pi

# Define metrics and dimensions to compare
metrics = ['ndcg@10', 'mrr@10', 'map@100', 'accuracy@1']
dims_to_compare = [64, 512] # Comparing Low vs High resource

fig, axes = plt.subplots(1, 2, figsize=(15, 8), subplot_kw=dict(polar=True))
fig.suptitle('Model Robustness: Low (64d) vs High (512d) Dimensions', fontsize=18, fontweight='bold')

for idx, dim in enumerate(dims_to_compare):
    ax = axes[idx]
    
    # Angles for the radar
    N = len(metrics)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1] # Close the loop
    
    # Helper to plot one model
    def plot_radar(model_name, data_source, color):
        if dim not in data_source: return
        values = [data_source[dim][m] for m in metrics]
        values += values[:1] # Close the loop
        ax.plot(angles, values, linewidth=2, linestyle='solid', label=model_name, color=color)
        ax.fill(angles, values, color=color, alpha=0.15)

    plot_radar("MPNet", mpnet_data, "#2ca02c")
    plot_radar("Shihab17", shihab_data, "#1f77b4")
    plot_radar("Distiluse", distiluse_data, "#d62728")

    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(metrics, fontsize=12, fontweight='bold')
    ax.set_title(f'Dimension: {dim}d', size=15, pad=20)
    
    # Set fixed y-limit to make comparisons fair
    ax.set_ylim(0, 0.7) 

# Legend only on the first plot to save space
axes[0].legend(loc='upper right', bbox_to_anchor=(0.1, 1.1))

plt.tight_layout()
plt.savefig("paper_plot_radar_chart.png", dpi=300)
plt.show()
```

```python
# Calculate 'Gap to Best' (assuming MPNet is best)
# We will create a dataframe showing the Difference compared to MPNet at the same dimension
gap_data = []

for dim in [64, 128, 256, 512]: # Common dimensions
    base_score = mpnet_data[dim]['ndcg@10'] # Using NDCG as the primary representative metric
    
    gap_data.append({
        "Dimension": dim, 
        "Model": "Shihab17", 
        "Gap": (shihab_data[dim]['ndcg@10'] - base_score) * 100 # Percentage point diff
    })
    gap_data.append({
        "Dimension": dim, 
        "Model": "Distiluse", 
        "Gap": (distiluse_data[dim]['ndcg@10'] - base_score) * 100
    })
    gap_data.append({
        "Dimension": dim, 
        "Model": "MPNet", 
        "Gap": 0.0 # Baseline
    })

df_gap = pd.DataFrame(gap_data)

plt.figure(figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Plot
ax = sns.barplot(x="Dimension", y="Gap", hue="Model", data=df_gap, 
                 palette=custom_palette, hue_order=["MPNet", "Shihab17", "Distiluse"])

# Add a zero line
ax.axhline(0, color="black", linewidth=1)

ax.set_title("Performance Gap relative to MPNet (NDCG@10)", fontsize=16)
ax.set_ylabel("Difference in Score (Percentage Points)", fontsize=12)
ax.set_xlabel("Embedding Dimension", fontsize=12)
ax.legend(title="Model", loc="lower left")

# Annotate bars
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', padding=3)

plt.tight_layout()
plt.savefig("paper_plot_relative_gap.png", dpi=300)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# --- Data Preparation (Using the data defined in previous steps) ---
# Re-declaring for standalone execution if needed
shihab_data = {
    64:  {'ndcg@10': 0.4783}, 128: {'ndcg@10': 0.5268},
    256: {'ndcg@10': 0.5510}, 512: {'ndcg@10': 0.5594}, 768: {'ndcg@10': 0.5638}
}
distiluse_data = {
    64:  {'ndcg@10': 0.3662}, 128: {'ndcg@10': 0.4004},
    256: {'ndcg@10': 0.4135}, 512: {'ndcg@10': 0.4213}
}
mpnet_data = {
    64:  {'ndcg@10': 0.5282}, 128: {'ndcg@10': 0.5783},
    256: {'ndcg@10': 0.6000}, 512: {'ndcg@10': 0.6110}, 768: {'ndcg@10': 0.6139}
}

# Calculate 'Performance Retention %' relative to MPNet
retention_data = []
target_dims = [64, 128, 256, 512] # Distiluse stops at 512

for dim in target_dims:
    base_score = mpnet_data[dim]['ndcg@10']
    
    # Shihab calculation
    retention_data.append({
        "Dimension": dim, "Model": "Shihab17", 
        "Retention": (shihab_data[dim]['ndcg@10'] / base_score) * 100
    })
    # Distiluse calculation
    retention_data.append({
        "Dimension": dim, "Model": "Distiluse", 
        "Retention": (distiluse_data[dim]['ndcg@10'] / base_score) * 100
    })
    # MPNet calculation (Always 100%)
    retention_data.append({
        "Dimension": dim, "Model": "MPNet", 
        "Retention": 100.0
    })

df_retention = pd.DataFrame(retention_data)

# --- Plotting ---
sns.set_theme(style="whitegrid")
custom_palette = {"Shihab17": "#1f77b4", "Distiluse": "#d62728", "MPNet": "#2ca02c"}

plt.figure(figsize=(12, 7))

# Create Bar Plot
ax = sns.barplot(
    x="Dimension", 
    y="Retention", 
    hue="Model", 
    data=df_retention, 
    palette=custom_palette, 
    hue_order=["MPNet", "Shihab17", "Distiluse"]
)

# Visual Enhancements
ax.set_title("Performance Retention Relative to MPNet (NDCG@10)", fontsize=18, fontweight='bold')
ax.set_ylabel("Performance Retention (%)", fontsize=14)
ax.set_xlabel("Embedding Dimension", fontsize=14)
ax.set_ylim(60, 105) # Zoom in to show the differences clearly (starts at 60%)

# Add a dashed line at 95% to show the "success threshold"
ax.axhline(95, color='gray', linestyle='--', alpha=0.5)
ax.text(3.6, 95.5, '95% Threshold', color='gray', fontsize=10, fontweight='bold')

# Annotate bars with specific percentages
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', padding=3, fontsize=11)

plt.legend(title="Model", loc="lower left", fontsize=12)
plt.tight_layout()

# Save
plt.savefig("paper_plot_retention.png", dpi=300)
plt.show()
```
